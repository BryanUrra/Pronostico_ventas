###  <font color=#CA0500> 1.5 Almacenar valores en una lista para resumir la limpieza de datos</font> 

```python
limpieza_datos = [] 

cant_stores= str(stores["Store"].count())
cant_feature= str(feature["Store"].count())
cant_sales= str(sales["Store"].count())

limpieza_datos.append(['DF_Stores',cant_stores])
limpieza_datos.append(['DF_Feature',cant_feature])
limpieza_datos.append(['DF_Weekly_Sales',cant_sales])
```

##  <font color=#F03415> 2. Limpieza de datos </font> 

###  <font color=#CA0500> 2.1 Tiendas tipo B con tamaño mayor a 100.000</font> 

```python
tiendas_filtrado = stores.loc[(stores["Type"] == "B") & (stores["Size"] > 100000)]
tiendas_filtrado.reset_index(inplace=True, drop=False)
del(tiendas_filtrado['index'])

cant_stores_filtrado= str(tiendas_filtrado["Store"].count())
limpieza_datos.append(['DF_Stores: TiendasB_tamaño>100.000',cant_stores_filtrado])

print("El tamaño mínimo de las tiendas tipo B del presente estudio es de: "
      + str(tiendas_filtrado['Size'].min()) + " lo cual es mayor a 100.000")
tiendas_filtrado
```

###  <font color=#CA0500> 2.2 Unir DF_Feature y DF_Stores, junto con filtrar por semanas con temperaturas mayores a 25° </font> 

```python
unir_df = feature.merge(tiendas_filtrado, on=['Store'], how='inner')

df_filtrado = unir_df.loc[(unir_df["Temperature"] > 25 )] 
df_filtrado.reset_index(inplace=True, drop=False)
del(df_filtrado['index'])

cant_feature_filtrado= str(df_filtrado["Store"].count())
limpieza_datos.append(['DF_filtrado (DF_Feature & DF_Stores): TiendaB_tam>100K_Temp>25',cant_feature_filtrado])

print("La temperatura mínima de cada semana es de: "+ str(df_filtrado['Temperature'].min()) + "° lo cual es mayor a 25°")
```

###  <font color=#CA0500> 2.3 Unir todos los dataframes considerando las restricciones iniciales</font> 

Se realiza un inner join considerando como llave 'Store', 'Date', 'IsHoliday', ya que se compara previamente la cantidad de 
valores únicos entre cada dataframe utilizando las siguientes lineas de códigos:
- sales['Date'].unique() 
- sales['IsHoliday'].unique()
- df_filtrado['Date'].unique()
- df_filtrado['IsHoliday'].unique()

```python
universo = sales.merge(df_filtrado, on=['Store','Date','IsHoliday'], how='inner')
universo.reset_index(inplace=True, drop=False)
del(universo['index'])

cant_dataframes= str(universo["Store"].count())
limpieza_datos.append(['Unión de dataframes (DF_Feature & DF_Stores & DF_Weekly_Sales)',cant_dataframes])

universo.head(5)
```

###  <font color=#CA0500> 2.4 Análisis de datos </font> 

```python
universo.info()
```

###  <font color=#450500> 2.4.1 Modificar el formato de la fecha a datetime </font>

```python
#La columna "Date" es tipo objeto, por lo que se debe transformar a datetime
universo['Date'] =  pd.to_datetime(universo['Date'])

# Separar la fecha en años, meses y días para utilizarlos posteriormente.
universo['Año'] = universo['Date'].dt.year
universo['Mes'] = universo['Date'].dt.month
universo['Semana'] = universo['Date'].dt.week
```

###  <font color=#450500> 2.4.2 Modificar fechas festivas a valores númericos </font>

```python
#Cambiar columna IsHoliday a valores númericos
def cambiar_bool(i):
    if i == False:
        return 0
    return 1

camb_bool = universo['IsHoliday'].map(cambiar_bool)
universo['IsHoliday'] = camb_bool

#Ordenar el dataframe
universo= universo.filter(['Store', 'Type', 'Dept' ,'Size', 'Weekly_Sales','Temperature','Fuel_Price','CPI',
                           'IsHoliday','Date','Año','Mes','Semana','Unemployment','MarkDown1','MarkDown2',
                           'MarkDown3','MarkDown4','MarkDown5'])

universo.head(5)
```

###  <font color=#450500> 2.4.3 Simplificar columnas del dataframe </font>

```python
print("La cantidad de registros en el dataframe es de : " +str(cant_dataframes))
print("---------------------------------------------------")
print("Datos nulos en las columnas del dataframe filtrado:")
universo.isnull().sum()
```

```python
#Eliminar columnas MarkDown dado que más de la mitad de los datos se encuentran con valores nulos, lo que podría generar un
#sesgo de la información si reeemplazamos los valores a cero. Además, dado que sólo se utilizará los tipos de 
#departamentos B (filtrados anteriormente), se elimina la columna "Type"

universo = universo.drop(['Type','MarkDown1','MarkDown2','MarkDown3','MarkDown4','MarkDown5'], axis=1)
```

###  <font color=#450500> 2.4.4 Outliers </font>

```python
print("Cantidad de ventas negativas: "+ str(len(universo["Weekly_Sales"][universo.Weekly_Sales < 0])))
print("Cantidad sin ventas: "+ str(len(universo["Weekly_Sales"][universo.Weekly_Sales == 0])))
```

```python
ventas_neg = universo["Weekly_Sales"][universo.Weekly_Sales < 0]
sns.displot(ventas_neg,height=6,aspect=3)
plt.title("Ventas negativas", fontsize=15)
plt.show()
```

```python
#Mantener ventas positivas o ceros
universo=universo[universo['Weekly_Sales']>=0]
```

```python
validar_datos = universo.groupby(['Store', 'Dept']).Weekly_Sales.agg(['max', 'min', 'mean', 'median', 'std']).reset_index()
validar_datos
```

```python
cant_dataframes_outliers= str(universo["Store"].count())
limpieza_datos.append(['Universo de datos',cant_dataframes_outliers])
```

```python
for i in universo:
    if universo[i].dtypes == float:
        universo[i]=universo[i].astype('int64')
```

```python
adt = seasonal_decompose(desc, model='aditive') 
plt.rc("figure",figsize=(12,8))
adt.plot()
plt.show()
```

###  <font color=#CA0500> 3.3 Datos de entrenamiento y prueba</font> 

```python
universo.sort_values(by=['Date'],inplace=True)
universo.set_index(universo.Date, inplace=True)
universo.head(5)
```

El modelo se entrenará con el 80% del universo en estudio y se probará con el 20% de los datos restantes. Cabe mencionar que, la función 'train_test_split' permite hacer una división de un conjunto de datos en dos bloques de entrenamiento y prueba de un modelo (train and test).

```python
from sklearn.model_selection import train_test_split 

uni_X=universo.drop(['Weekly_Sales','Date'], axis=1)
uni_Y=universo['Weekly_Sales'] 

X_train,X_test,y_train,y_test = train_test_split(uni_X,uni_Y,test_size=0.20, random_state=10)

X_tr_rl,X_tes_rl,y_tr_rl,y_tes_rl = X_train.copy(),X_test.copy(),y_train.copy(),y_test.copy()
X_tr_rf,X_tes_rf,y_tr_rf,y_tes_rf = X_train.copy(),X_test.copy(),y_train.copy(),y_test.copy()
X_tr_ad,X_tes_ad,y_tr_ad,y_tes_ad = X_train.copy(),X_test.copy(),y_train.copy(),y_test.copy()
X_tr_kn,X_tes_kn,y_tr_kn,y_tes_kn = X_train.copy(),X_test.copy(),y_train.copy(),y_test.copy()
X_tr_xgbr,X_tes_xgbr,y_tr_xgbr,y_tes_xgbr = X_train.copy(),X_test.copy(),y_train.copy(),y_test.copy()
```

##  <font color=#F03415> 4. Modelos de machine learning </font> 

###  <font color=#CA0500> 4.1 Librerías para los algoritmos de machine learning </font> 

```python
from sklearn import metrics
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
```

###  <font color=#CA0500> 4.2 Regresión lineal </font> 

```python
lr = LinearRegression(normalize=False)
lr.fit(X_tr_rl, y_tr_rl)
```

```python
lr_acc = round(lr.score(X_tes_rl,y_tes_rl)*100,2)
print("Precisión de la regresión lineal: "+ str(lr_acc) + "% ")
```

```python
y_pred_rl = lr.predict(X_tes_rl)
```

```python
print("MAE" , metrics.mean_absolute_error(y_tes_rl, y_pred_rl))
print("MSE" , metrics.mean_squared_error(y_tes_rl, y_pred_rl))
print("RMSE" , np.sqrt(metrics.mean_squared_error(y_tes_rl, y_pred_rl)))
print("R2" , metrics.explained_variance_score(y_tes_rl, y_pred_rl))
```

```python
plt.figure(figsize=(20,8))
plt.plot(lr.predict(X_tes_rl[:200]), label="predicción", linewidth=2.0,color='blue')
plt.plot(y_tes_rl[:200].values, label="valores reales", linewidth=2.0,color='red')
plt.legend(loc="best")
plt.title("Pronóstico con regresión lineal")
plt.show()
```
